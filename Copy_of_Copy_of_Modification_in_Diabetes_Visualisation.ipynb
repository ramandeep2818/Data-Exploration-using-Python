# Analysis of the project with graphical representation

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dt.head()
```

* **Scatter Plot** :
A scatter plot graph uses two axes to show how two continuous factors are related and spread out. Each data point is placed along one of the axes. It shows patterns, correlations, or clusters in data visually, which helps you figure out how strong the link is and which way the trend is going between the factors being shown. The spread of dots across the graph shows how the factors being looked at can change and how they might be related.



```python
# Scatter plot
sns.scatterplot(x='DP Function', y='Serum Ins', data= dt)
plt.title('DP Function vs Serum Ins')
plt.show()

```

* **Histogram**:A histogram is a graph that shows the count or number of values in each bin of numerical data by dividing it into intervals known as bins. It shows the shape, central tendency, and dispersion of the data graphically, which helps you understand how the data is spread out. Histograms help you find trends and outliers in a set of data and also help you figure out how the data is distributed overall.


```python
# Histogram
sns.histplot(data=dt, x='Diastolic BP', bins=20, color='yellow', alpha=0.9, label='DP Function')
sns.histplot(data=dt, x='Serum Ins', bins=20, color='green', alpha=0.9, label='Serum Ins')
plt.title('Histogram of DP Function and Serum Ins')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()


```

* **Box Plot** : A box plot, also called a box-and-whisker map, shows how numbers are spread out by quartiles and draws attention to the median, interquartile range (IQR), and outliers. It shows the data's mean, variance, and skewness, which makes it easy to quickly compare groups or datasets. There is a clear visual description of how variable the data is in box plots, which show dispersion and the presence of extreme numbers.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Outcome' is the target variable (0 for non-diabetic, 1 for diabetic)
plt.figure(figsize=(10, 6))
sns.violinplot(x='Tri Fold Thick', y='Age', data=dt, inner=None, color='lightgray')
sns.swarmplot(x='Tri Fold Thick', y='Age', data=dt, color='purple', alpha=0.7)
plt.title('Age Distribution by Diabetes Tri Fold Thick')
plt.xlabel('Tri_Fold_Thick (0: Non-diabetic, 1: Diabetic)')
plt.ylabel('Age Level')
plt.show()
```

```python
# Box plot
sns.boxplot(data=dt[['Age', 'Diastolic BP']])
plt.title('Box Plot of Age and Diastolic BP')
plt.show()
sns.boxplot(data=dt[['Serum Ins','Diastolic BP']])
plt.title('Box Plot of Serum Ins and Diastolic BP')
plt.show()
sns.boxplot(data=dt[['BMI','Age','Pregnancies']])
plt.title('Box Plot of Age and feature2')
plt.show()
```

## Corelation Matrix

A correlation matrix is a table that shows the pairwise correlations between variables in a dataset. The numbers in the matrix range from -1 to 1, which show how strong and which way the relationships are going. It lets you find connections between different variables, which helps you figure out how changes in one variable affect changes in others. The matrix shows relationships in a useful way, which helps choose features, check for multicollinearity, and find important factors that affect data analysis or prediction models.


```python
correlation_matrix = dt.corr()
plt.figure(figsize=(5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of classes')
plt.show()
```

## Pairplot Graph repesent

Pairplot graphs, made possible by the seaborn module in Python, use scatterplots and histograms to show the pairwise correlations between numerical variables in a dataset. The data is presented visually in a grid of graphs that compare each variable to each other. This allows for quick understanding of distributions, correlations, and likely patterns. When doing initial exploratory data analysis, pairplot graphs are useful because they reveal the interrelationships and correlations between the variables.



```python
selected_features = ['DP Function', 'Serum Ins', 'Tri Fold Thick', 'class']
sns.pairplot(dt[selected_features], hue='class', markers=['o', 's'])
plt.suptitle('Pairplot Graph of classes', y=1.00, ha='center', fontsize='large')
plt.show()
```

# Bar Graph

```python
X = dt.drop('class', axis=1)
y = dt['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='pie')
plt.title('Feature Importances')
plt.show()
```
